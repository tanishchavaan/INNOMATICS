Unique Words in a Sentence



```python
sentence=input("Enter the sentence :")
words=sentence.split()
unique=set(words)
print("The number of unique words are ",len(unique))
print("The unique words are ",unique)


```

Highest Salary from Employee Data


```python
# Employee dictionary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Finding employee with highest salary
highest_paid = max(employees, key=employees.get)

# Printing result
print("Highest Salary:", highest_paid, "-", employees[highest_paid])
```

Count Products Above a Price Threshold



```python
prices = [450, 1200, 899, 1500, 300]
count=0
for i in prices :
  if i>1000:
    count+=1
print("The numbers greater than 1000 : ",count)
```

Calculate Attendance Percentage


```python
attendance = ["P", "P", "A", "P", "P"]
total=len(attendance)
count=0

for i in attendance:
  if i=='P':
    count+=1
percent=(count/total)*100
print("The perecentage of present is :",percent)
```

 Count Character Frequency



```python
text = "python"

char_count = {}

for ch in text:
    if ch in char_count:
        char_count[ch] += 1
    else:
        char_count[ch] = 1

print(char_count)
```
