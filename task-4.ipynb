Mobile Recharge Validation System


```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        print("Minimum recharge amount is ₹50.")
        return False
    elif amount not in valid_plans:
        print("Invalid plan selected.")
        print("Available plans: 199, 299, 399, 599")
        return False
    else:
        print(" Recharge Successful!")
        return True


# Retry system using while loop
while True:
    try:
        user_amount = int(input("Enter recharge amount: ₹"))

        if validate_recharge(user_amount):
            break

    except ValueError:
        print("Please enter a valid numeric amount.")

```

 Inventory Reorder Alert System


```python
def check_stock(products):
    for product, quantity in products.items():
        if quantity < 15:
            print(f" {product}: Reorder Alert (Stock: {quantity})")
        else:
            print(f"{product}: Stock OK (Stock: {quantity})")

inventory = {
    "Rice": 20,
    "Sugar": 10,
    "Milk": 5,
    "Oil": 25
}

check_stock(inventory)
```

Student Result Processing System


```python
def evaluate_student(marks):
    total = 0

    for mark in marks:
        total += mark

    average = total / len(marks)

    if average >= 50:
        return f"Average: {average:.2f} → Pass "
    else:
        return f"Average: {average:.2f} → Fail "

student_marks = [60, 45, 70, 55, 40]

result = evaluate_student(student_marks)
print(result)
```

Cab Fare Estimator with Retry Option


```python
def calculate_fare(distance, peak):
    base_fare = 50
    per_km_rate = 12

    total_fare = base_fare + (distance * per_km_rate)

    if peak:
        total_fare += total_fare * 0.25

    return total_fare



while True:
    try:
        distance = float(input("Enter distance in km: "))
        peak_input = input("Is it peak hour? (yes/no): ").lower()

        if peak_input not in ["yes", "no"]:
            print("Please enter 'yes' or 'no'")
            continue

        peak = True if peak_input == "yes" else False

        fare = calculate_fare(distance, peak)
        print(f"Total Fare: ₹{fare:.2f}")

        break

    except ValueError:
        print("Please enter valid numeric distance.")
```

Employee Attendance Eligibility Checker


```python
def check_attendance(attendance_list):
    present_count = 0


    for status in attendance_list:
        if status == "P":
            present_count += 1

    total_days = len(attendance_list)

    if total_days == 0:
        return "No attendance data available."

    percentage = (present_count / total_days) * 100

    if percentage >= 75:
        return f"Attendance: {percentage:.2f}% → Eligible "
    else:
        return f"Attendance: {percentage:.2f}% → Not Eligible "



attendance = ["P", "P", "A", "P", "A", "P", "P", "P"]

result = check_attendance(attendance)
print(result)
```
