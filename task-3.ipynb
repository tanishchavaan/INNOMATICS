1: Employee Performance Bonus Eligibility


```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

highest_score = max(employees.values())
top_performers = [name for name, score in employees.items() if score == highest_score]
print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {highest_score})")
```

2: Search Query Keyword Analysis


```python
query = "Buy mobile phone buy phone online"

#lowercase
query = query.lower()

#  Remove punctuation manually
cleaned_query = ""
for char in query:
    if char.isalnum() or char.isspace():
        cleaned_query += char

# Split into words
words = cleaned_query.split()

#  Count frequency
frequency = {}

for word in words:
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1

# Filter words searched more than once
result = {}
for word, count in frequency.items():
    if count > 1:
        result[word] = count

print(result)
```

3: Sensor Data Validation


```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for index in range(len(sensor_readings)):
    if sensor_readings[index] % 2 == 0:   # check if even
        valid_readings.append((index, sensor_readings[index]))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

4: Email Domain Usage Analysis


```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

#  Extract and count domains
for email in emails:
    domain = email.split("@")[1]

    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

#  Calculate percentage
total_users = len(emails)

for domain, count in domain_count.items():
    percentage = (count / total_users) * 100
    print(f"{domain}: {int(percentage)}%")
```

 5: Sales Spike Detection


```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average
avg = sum(sales) / len(sales)

# Detect spike days
for index, value in enumerate(sales):
    increase_percentage = ((value - avg) / avg) * 100

    if increase_percentage >= 29:   # Approx 30% spike
        print(f"Day {index + 1}: {value}")
```

6: Duplicate User ID Detection


```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

frequency = {}

# Count occurrences
for user in user_ids:
    if user in frequency:
        frequency[user] += 1
    else:
        frequency[user] = 1

# Display duplicates
for user, count in frequency.items():
    if count > 1:
        print(f"{user} → {count} times")
```
