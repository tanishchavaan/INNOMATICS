 1 . Smart Parking Lot Management System


```python
def smart_parking_system(capacity, logs):
    current = 0
    peak = 0

    for log in logs:
        if log == "IN":
            current += 1
        elif log == "OUT":
            if current > 0:
                current -= 1


        if current > peak:
            peak = current


    if current > capacity:
        status = "Alert! Parking Exceeded Capacity"
    elif current == capacity:
        status = "Full"
    else:
        status = "Available"

    print("Currently Parked Vehicles:", current)
    print("Peak Parking Usage:", peak)
    print("Parking Status:", status)
capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking_system(capacity, logs)
```

2. Online Food Delivery Time Estimator


```python
def estimate_delivery_time(distance, traffic, weather):
    speed = 30  # km/hr

    # Base time calculation
    base_time = (distance / speed) * 60

    # Traffic adjustment
    if traffic == "Low":
        base_time += 5
    elif traffic == "Medium":
        base_time += 10
    elif traffic == "High":
        base_time += 20

    # Weather adjustment
    if weather == "Cloudy":
        base_time += 5
    elif weather == "Rainy":
        base_time += 15
    elif weather == "Storm":
        base_time += 25

    return round(base_time)
distance = 8
traffic = "High"
weather = "Rainy"

print("Estimated Delivery Time:", estimate_delivery_time(distance, traffic, weather), "minutes")
```

3. Movie Theatre Seat Occupancy

```python
def theatre_occupancy_analyzer(total_seats, booked_seats):
    booked_count = len(booked_seats)

    occupancy = (booked_count / total_seats) * 100

    # Determine show status
    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    elif occupancy >= 40:
        status = "Moderate"
    else:
        status = "Seats Available"

    # Suggest additional show
    if occupancy >= 90:
        suggestion = "Consider opening additional show"
    else:
        suggestion = "No extra show needed"

    print("Occupancy:", round(occupancy), "%")
    print("Show Status:", status)
    print("Suggestion:", suggestion)
total_seats = 200
booked_seats = [1]*150   # 150 booked seats

theatre_occupancy_analyzer(total_seats, booked_seats)
```

4. Cloud Server Load Classification System


```python
def classify_server_load(cpu_readings):
    total = 0

    # Loop to calculate sum
    for reading in cpu_readings:
        total += reading

    average = total / len(cpu_readings)

    # Classification
    if average < 50:
        status = "Normal"
    elif average <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print("Average CPU Load:", round(average), "%")
    print("Server Status:", status)
cpu = [45, 60, 70, 85, 90]

classify_server_load(cpu)
```

5.  Smart Classroom Resource Usage Monitor


```python
def event_registration_controller(capacity, registrations):
    confirmed = 0
    waitlisted = 0

    # Loop simulation (like users registering one by one)
    for i in range(registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1

    # Registration Status
    if confirmed >= capacity:
        status = "Closed"
    else:
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)
capacity = 100
registrations = 105

event_registration_controller(capacity, registrations)
```
