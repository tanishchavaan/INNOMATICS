2. Pass / Fail Analyzer
**bold text**


```python
marks = [45, 78, 90, 33, 60]
pass_count= 0
fail_count= 0
for i in marks :
  if i>49:
    pass_count+=1 #pass count increases if marks is greater than 49
  else:
    fail_count+=1 #fail count increases if marks is less than 49
print("Total students :"+str(len(marks)))
print("The number of students passed :"+str(pass_count)) #converting the numbers to strings for concatenation
print("The number of students failed :"+str(fail_count))
```

3. **Simple Data Cleaner**

```python
names = [" Alice ", "bob", " CHARLIE "]
cleaned_names=[]
for i in names:
  cl=i.strip().lower() #stip removes white spaces and lower makes it all lowercase
  cleaned_names.append(cl)  #all the new items are appended or added to a new list
print("The new list is ",cleaned_names)
```

4. **Message Length Analyzer**


```python
messages = ["Hi", "Welcome to the platform", "OK"]
for m in messages :
  print("message:",m)
  l=len(m)
  print("The length of msg is ",l)
  if(l>10):  #flagging the msg with more than 10 characters
    print("Flag as the length of msgs is geater than 10")

  print()
```

5. **Error Message Detector**


```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
count=0 #initializing count as 0
for i in logs:
  if i=="ERROR":
    count+=1 #counting each time error is seen

print("the number of error entries : ",count)
```
