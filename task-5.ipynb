 1. Social Media – Post Engagement Analyzer

```python
def analyze_post_engagement(likes_list):
    total_likes = 0

    # Loop to calculate total likes
    for likes in likes_list:
        total_likes += likes

    # Check post status
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    # Output
    print(f"Total Likes: {total_likes}")
    print(f"Post Status: {status}")


# Example usage
likes_data = [200, 300, 150, 400]
analyze_post_engagement(likes_data)
```

 3. Agriculture – Rainfall Adequacy Checker


```python
def analyze_rainfall(rainfall_data, required_level):

    total_rainfall = 0

    # Loop to calculate total rainfall
    for rain in rainfall_data:
        total_rainfall += rain

    # Calculate average
    average_rainfall = total_rainfall / len(rainfall_data)

    # Check rainfall status
    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    # Output
    print(f"Average Rainfall: {int(average_rainfall)}")
    print(f"Rainfall Status: {status}")


# Example usage
season_data = [60, 80, 70, 78]
analyze_rainfall(season_data, 70)
```

4. Social Media – Duplicate Account Detection


```python
def detect_duplicate_usernames(usernames):

    unique_usernames = set(usernames)

    if len(unique_usernames) < len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


# Example usage
user_list = ["alex", "john", "maria", "alex"]
detect_duplicate_usernames(user_list)
```

6. Agriculture – Premium Crop Price Filter


```python
def filter_premium_crops(prices):

    premium_crops = []

    for price in prices:
        if price > 2000:
            premium_crops.append(price)

    print(f"Premium Crops: {premium_crops}")


# Example usage
crop_prices = [1500, 2500, 1800, 3200]
filter_premium_crops(crop_prices)
```

```python
def evaluate_system_health(errors):

    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print(f"Error Count: {errors}")
    print(f"System Status: {status}")


# Example usage
evaluate_system_health(7)
```

9. E-Learning – Student Attendance Eligibility System


```python
def check_exam_eligibility(attendance_list):
    total_days = len(attendance_list)
    present_days = 0

    for day in attendance_list:
        if day == 1:
            present_days += 1

    attendance_percentage = (present_days / total_days) * 100

    if attendance_percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print(f"Attendance Percentage: {attendance_percentage}")
    print(f"Exam Eligibility: {status}")


# 1 = Present, 0 = Absent
attendance_data = [1, 1, 1, 1, 0]  # 4 present out of 5

check_exam_eligibility(attendance_data)
```

10. Smart Electricity Bill Analyzer


```python
def calculate_electricity_bill(units):
    bill = 0

    # Slab 1: First 100 units
    if units <= 100:
        bill = units * 3

    # Slab 2: 101–200 units
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)

    # Slab 3: Above 200 units
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage Classification
    if bill < 500:
        usage = "Low Usage"
    elif 500 <= bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    return bill, usage


# Example
units_consumed = 250
total_bill, usage_status = calculate_electricity_bill(units_consumed)

print(f"Units Consumed: {units_consumed}")
print(f"Total Bill: ₹{total_bill}")
print(f"Usage Classification: {usage_status}")
```
